```python
# importing the dependencies
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
# specification for dataset
X,y = make_classification(
    n_samples = 100,
    n_features = 2,
    n_informative = 1,
    n_redundant = 0,
    n_classes = 2,
    n_clusters_per_class = 1,
    random_state = 41,
    hypercube = False,
    class_sep = 10
)
```

```python
# ploting
plt.figure(figsize = (10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='bwr',s=100)
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis = 1)   # This line inserts a column of 1s at index 0 (the first column) of X.
  weight = np.ones(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weight))
    weight = weight + lr*(y[j] - y_hat)*X[j]

  return weight[0],weight[1:]
```

```python
# calling of function
def step(z):
  return 1 if z>0 else 0
```

```python
# calling of function
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
print(m)
print(b)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
